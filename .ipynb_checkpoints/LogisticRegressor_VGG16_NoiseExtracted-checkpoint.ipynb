```python
df_sample = df.sample(frac=1.0, random_state=0)
```

```python
X = df_sample.drop('class', axis=1)
stnd = StandardScaler()
stnd.fit(X)
X = stnd.transform(X)
y = df_sample['class']
```

```python
time_start = time()
lgr = LogisticRegression(n_jobs=-1,
                        max_iter=10000)

lgr.fit(X,y)

time_stop = time()
print('Elapsed Time for Logistic Regression:', time_stop-time_start)
print()
print('Baseline Training Accuracy:', lgr.score(X,y))
```

```python
time_start = time()
lgrcv = LogisticRegressionCV(cv=5, 
                             random_state=0, 
                             n_jobs=-1,
                             penalty='l1',
                             solver = 'saga',
                             max_iter=5000)

lgrcv.fit(X,y)
time_stop = time()
print('CV Time:', time_stop-time_start)
print()
print('CV Valid Error:', lgrcv.score(X,y))
```
